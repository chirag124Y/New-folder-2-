```python
                    # FOR LOOP
# Q.1
print("Armstrong numbers between 100 and 999 are:")
for num in range(100, 1000):
    digits = [int(d) for d in str(num)]
    sum_of_cubes = sum(d**3 for d in digits)

    if num == sum_of_cubes:
        print(num)
```

```python
# Q.3
def sum_of_digits(num):
    total = 0
    for digit in str(num):
        total += int(digit)
    return total

for number in range(1, 501):
    if number % 3 == 0:
        if sum_of_digits(number) <= 10:
            print(number, end=' ')
```

```python
# Q.5
text = input("Enter a string: ")
text = text.lower()  
alphabet = "abcdefghijklmnopqrstuvwxyz"
found = ""

for ch in text:               
    if ch in alphabet and ch not in found:
        found += ch

if len(found) == 26:
    print("The string is a pangram.")
else:
    print("The string is NOT a pangram.")
```

```python
            # WHILE LOOPS
# Q.1
num = int(input("Enter a number: "))
rev = 0
temp = num
while temp > 0:
    rev = rev * 10 + temp % 10
    temp //= 10
is_prime = True
if rev < 2:
    is_prime = False
else:
    for i in range(2, rev):
        if rev % i == 0:
            is_prime = False
            break
print(f"Reverse = {rev}")
print("Prime" if is_prime else "Not Prime")
```

```python
# Q.5
num = int(input("Enter a number: "))
i = 2
largest = 0
while i <= num:
    if num % i == 0:
        is_prime = True
        for j in range(2, i):
            if i % j == 0:
                is_prime = False
                break
        if is_prime:
            largest = i
        num //= i
    else:
        i += 1
print("Largest prime factor:", largest)
```

```python
# Q.9
num = int(input("Enter a number: "))
sq = str(num ** 2)
while True:
    for i in range(1, len(sq)):
        left = int(sq[:i]) if sq[:i] else 0
        right = int(sq[i:]) if sq[i:] else 0
        if left + right == num:
            print("Kaprekar number")
            break
    else:
        print("Not a Kaprekar number")
    break
```

```python
# Q.10
balance = 0
while True:
    print("\n1. Check Balance\n2. Deposit\n3. Withdraw\n4. Exit")
    choice = input("Choose an option: ")
    if choice == '1':
        print("Balance:", balance)
    elif choice == '2':
        amt = int(input("Enter amount to deposit: "))
        balance += amt
    elif choice == '3':
        amt = int(input("Enter amount to withdraw: "))
        if amt <= balance:
            balance -= amt
            print("Withdrawal successful.")
        else:
            print("Insufficient balance.")
    elif choice == '4':
        print("Exiting ATM.")
        break
    else:
        print("Invalid option.")
```
